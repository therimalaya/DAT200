# Import neede modules

```python
import hoggorm as ho
import hoggormplot as hopl
import pandas as pd
```

# Load data and extract information

```python
cancer_men_df = pd.read_table("../data/Cancer_men_perc.txt", sep='\t', index_col=0)
countryNames = list(cancer_men_df.index)
cancerNames = list(cancer_men_df.columns)
cancerDataMen = cancer_men_df.values
```

# Compute PCA model

```python
cm_model = ho.nipalsPCA(
    arrX = cancerDataMen,
    numComp = 8,
    Xstand = False,
    cvType = ["loo"]
)
```

```python
hopl.plot(
    cm_model, 
    comp = [1, 2], 
    plots = ["scores", "loadings"], 
    objNames = countryNames,
    XvarNames = cancerNames,
    figsize = (10, 7)
)
```

```python
cumCalExplVar = cm_model.X_cumCalExplVar()
calExplVar = cm_model.X_calExplVar()
```

```python
cumCalExplVar
```
